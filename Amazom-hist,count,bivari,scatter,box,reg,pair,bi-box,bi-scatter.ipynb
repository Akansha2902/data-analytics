```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Amazon video prime DataFrame

```python
df=pd.read_csv("Downloads/amazon_prime_video_data.csv")
```

```python
df.dtypes
```

```python
plt.hist(df["release_year"])
plt.title("Histogram for Release year")
plt.xlabel("Year")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows that higher sale in 2012-2017 year as follow 2022 and lowest sale in 2006-2010 year range.

```python
plt.hist(df["views_millions"])
plt.title("Histogram for views millions")
plt.xlabel("views millions")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows that the higher views are in range 60-70 and 80-110 and lower views are in range 0-15 as follow 70-80.

```python
plt.hist(df["avg_watch_time_min"])
plt.title("Histogram for average watch time")
plt.xlabel("avg_watch_time_min")
plt.ylabel("Frequency")
plt.show()
```

The distribution of average watch time in minutes shows that the higher watch time is 20-35 as follow 135-150 and lowest average watch time is 90-100 and 130-140.

```python
plt.hist(df["likes_millions"])
plt.title("Histogram for likes millions")
plt.xlabel("likes millions")
plt.ylabel("Frequency")
plt.show()
```

The distribution of likes in millions shows that the higher likes in range 12-13.5 and lower is 10-12.

```python
plt.hist(df["revenue_millions"])
plt.title("Histogram for revenue millions")
plt.xlabel("revenue millions")
plt.ylabel("Frequency")
plt.show()
```

The distribution of revenue in millions shows that the higher revenue in range 35-40 as follow 25-30 and lower revenue in range 0-5.

## countplot

```python
sns.countplot(x=df["genre"],data=df)
plt.title("Count plot for genre")
plt.xlabel("genre")
plt.show()
```

The count plot shows that the highest count of genre is documentary and drama and lowest count of genre is comedy as follow romance movies.

```python
sns.countplot(x=df["language"],data=df)
plt.title("Count plot for language")
plt.xlabel("language")
plt.show()
```

The count plot shows that the highest count of languages is tamil as follow malayalam and lowest count of language is kannada as follow telugu.

```python
sns.countplot(x=df["age_rating"],data=df)
plt.title("Count plot for age_rating")
plt.xlabel("age_rating")
plt.show()
```

The count plot shows that the highest count of age rating is U and lowest count of age rating is U/A.

```python
sns.countplot(x=df["country"],data=df)
plt.title("Count plot for country")
plt.xlabel("country")
plt.show()
```

The count plot shows that the highest cout of country is Australia as follow UK and lowest count of country is USA as follow Canada.

```python
sns.countplot(x=df["subscription_plan"],data=df)
plt.title("Count plot for subscription plan")
plt.xlabel("subscription plan")
plt.show()
```

The count plot shows that the highest count of subscription plan is Basic and lowest count of subscription plan is Premium.

```python
sns.countplot(x=df["is_trending"],data=df)
plt.title("Count plot for trending")
plt.xlabel("trending")
plt.show()
```

The count plot shows that the highest count of not trending and lowest is trending movies in Amazon prime.

## bivariate 

```python
sns.countplot(x=df["genre"],hue="language",data=df)
plt.title("Genre wise language")
plt.xlabel("Genre")
plt.ylabel("Language")
plt.xticks(rotation=90)
plt.show()
```

The Bivariate plot shows that the highest genre as per language is documentary in hindi language as follow action in hindi language and lowest genre as per language is romance in kannada language.

```python
sns.countplot(x=df["age_rating"],hue="country",data=df)
plt.title("Age rating wise country")
plt.xlabel("Age rating")
plt.ylabel("Country")
plt.show()
```

The bivariate plot shows that the highest age rating as per country is U in Australia as follow canada and lowest age rating as per country is U in USA. 

```python
sns.countplot(x=df["subscription_plan"],hue="is_trending",data=df)
plt.title("Subscription plan wise trending")
plt.xlabel("Subscription plan")
plt.ylabel("Trending")
plt.show()
```

The bivariate plot shows that the highest subscription plan as per trending is Basic and lowest subscription plan as per trending is Standard.

## Reggration plot


```python
sns.regplot(x="release_year",y="revenue_millions",data=df)
plt.title("reggration plot for release year and revenue")
plt.show()
```

the reggration plot shows that the downward trend in revenue as per release year.That means revenue decreases as per low releasing movies as per up coming year.

```python
sns.regplot(x="imdb_rating",y="revenue_millions",data=df)
plt.title("reggration plot for imdb rating and revenue")
plt.show()
```

the reggration plot shows that the upward trend in revenue as per imdb rating.That means the imdb rating is increases then revenue also increases.

```python
sns.regplot(x="revenue_millions",y="views_millions",data=df)
plt.title("reggration plot for revenue and views")
plt.show()
```

the reggration plot shows that the downward trend in revenue as per views. that means revenue increases then views are decreases.

```python
sns.regplot(x="imdb_rating",y="views_millions",data=df)
plt.title("reggration plot for imdb rating and views")
plt.show()
```

the reggration plot shows that there is uniform trend in imdb rating as per viwes.

```python
sns.regplot(x="likes_millions",y="avg_watch_time_min",data=df)
plt.title("reggration plot for likes and average watch time in minutes")
plt.show()
```

the reggration plot shows that the downward trend in likes as per average watch time.That means when likes increases then average watch time is decreasing.
