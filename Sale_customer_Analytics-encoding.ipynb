```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("sales_customer_analytics.csv")
```

```python
df.head()
```

```python
city=df["City"]
encoder=LabelEncoder()
city_encoded=encoder.fit_transform(city)
print(city_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
product=df["Product"]
encoder=LabelEncoder()
product_encoded=encoder.fit_transform(product)
print(product_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
cat=df["Category"]
encoder=LabelEncoder()
cat_encoded=encoder.fit_transform(cat)
print(cat_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
payment=df["PaymentMode"]
encoder=LabelEncoder()
payment_encoded=encoder.fit_transform(payment)
print(payment_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
order=df["OrderStatus"]
encoder=LabelEncoder()
order_encoded=encoder.fit_transform(order)
print(order_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```
