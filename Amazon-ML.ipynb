## Label encoding

```python
gen=df["genre"]
encoder=LabelEncoder()
gen_encoded=encoder.fit_transform(gen)
print(gen_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
lang=df["language"]
encoder=LabelEncoder()
lang_encoded=encoder.fit_transform(lang)
print(lang_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
coun=df["country"]
encoder=LabelEncoder()
coun_encoded=encoder.fit_transform(coun)
print(coun_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

## One hot encoding

```python
df=pd.DataFrame({"gen":df["genre"]})
encoded_df=pd.get_dummies(df,columns=["gen"],dtype=int)
print(encoded_df)
```

```python
df1=df.copy()
```

```python
from sklearn.model_selection import train_test_split
x=df1.drop("is_trending",axis=1)
y=df1["is_trending"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
print("Training set size:",x_train.shape)
print("Test set size:",x_test.shape)
print("Churn distribution in Train:\n",y_train.value_counts(normalize=True))
print("Churn distribution in test:\n",y_test.value_counts(normalize=True))
```

```python
print(df1.columns)
```

```python
import pandas as pd
x_train = pd.get_dummies(x_train, drop_first=True)
x_test = pd.get_dummies(x_test, drop_first=True)
x_train, x_test = x_train.align(x_test, join='left', axis=1, fill_value=0)
```

```python
x_train = x_train.fillna("Unknown")
x_test = x_test.fillna("Unknown")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, ConfusionMatrixDisplay,accuracy_score
log_reg = LogisticRegression(max_iter=5000, solver='liblinear')
log_reg.fit(x_train, y_train)

y_pred = log_reg.predict(x_test)
y_prob = log_reg.predict_proba(x_test)[:,1]
```

```python
print("Accuracy Score:",accuracy_score(y_test,y_pred))
```

```python
print("Classification report:\n",classification_report(y_test,y_pred))
```

```python
ConfusionMatrixDisplay.from_estimator(log_reg,x_test,y_test,cmap="Blues")
```

```python
print("ROC-AUC score:",roc_auc_score(y_test,y_prob))
```

```python
y_train = y_train.map({'No':0, 'Yes':1})
y_test = y_test.map({'No':0, 'Yes':1})
```

```python
from sklearn.metrics import roc_curve

fpr,tpr,_=roc_curve(y_test,y_prob)
plt.plot(fpr,tpr,label="Logistic regression")
plt.plot([0,1],[0,1],'--',color="gray")
plt.xlabel("false positive rate")
plt.ylabel("true positive rate")
plt.title("ROC curve")
plt.legend()
plt.show()
```
