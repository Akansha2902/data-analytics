```python
'''1.Count how many lists are present within a nested list (e.g., [[1, 2], [3, 4], [12]] returns 3)'''
n= [[1, 2], [3, 4], [12]]
count = 0
for i in n:
    if type(i) == list:  
        count += 1
print(count)
```

```python
'''4.Remove duplicate lists from a list of lists (e.g., [[1,2],[3,4],[1,2],[5,6]] should become unique)'''
l= [[1,2], [3,4], [1,2], [5,6]]
unique = []
for i in l:
    if i not in unique:   
        unique.append(i)
print(unique)
```

```python
'''5.Find the length of the longest increasing subsequence in a list'''
n= [10, 22, 9, 33, 21, 50, 41, 60]
p = [1] * len(n)
for i in range(len(n)):
    for j in range(i):
        if n[j] < n[i]:        
            p[i] = max(p[i],p[j] + 1)
print(max(p))
```

```python
'''7.Merge multiple lists so that only unique elements remain, without using set comprehensions'''
l= [[1, 2, 3], [2, 4, 6], [1, 5, 6, 7]]
unique= []
for lst in l:
    for i in lst:
        if i not in unique:
            unique.append(i)
print(unique)
```

```python
'''8.Count the total number of elements (not sublists) inside a nested list regardless of depth'''
n= [1, [2, 3], [4, [5, 6, [7]]], 8]
m= n[:]  
total = 0
i = 0
while i < len(m):
    if type(m[i]) == list:
        m= m[:i] + m[i] + m[i+1:]
    else:
        total += 1
        i += 1 
print(total)
```

```python
'''9.Find all pairs of numbers within a list that sum to a specific target (e.g., 10)'''
n= [2, 4, 3, 5, 7, 1, 6, 8]
m= 10
pairs = []
for i in range(len(n)):
    for j in range(i + 1, len(n)):
        if n[i] + n[j] == m:
            pairs.append((n[i], n[j]))
print(pairs)
```

```python
'''10.Write a function that moves all zeroes in a list to the end while maintaining the order of other elements.'''
l= [0, 1, 0, 3, 12, 0, 5]
result = []
count= 0
for n in l:
    if n == 0:
        count+= 1      
    else:
        result.append(n)   
result.extend([0] * count)
print(l)
print(result)
```

```python
'''5.Write a python function to Write a function that finds the minimum and maximum of a tuple manually (without min() or max()).'''
def min_max(t):
    if len(t) == 0:
        return None, None  
    minimum = t[0]
    maximum = t[0]
    for i in t:
        if i < minimum:
            minimum = i
        if i> maximum:
            maximum = i
    return minimum, maximum
t = (3, 12, 8, 90, 11, 23, 89, 45, 66, 22)
min_val, max_val = min_max(t)
print(f"Minimum:{min_val}")
print(f"Maximum:{max_val}")
```

```python
'''7.Write a python function to Given a list of coordinate pairs as tuples, return all x-values and y-values as separate tuples.
'''
def coordinate(c):
    x_vals = tuple(x for x, y in c)
    y_vals = tuple(y for x, y in c)
    return x_vals, y_vals
coordinates = [(1, 2), (3, 4), (5, 6), (7, 8)]
x_tuple, y_tuple =coordinate(coordinates)
print("X-values:", x_tuple)
print("Y-values:", y_tuple)
```

```python
'''8.Write a python function to Find the most frequently occurring element in a tuple.
'''
def frequent(t):
    if not t:
        return None  
    max_count = 0
    most_freq = t[0]
    for i in t:
        count = 0
        for x in t:
            if x == i:
                count += 1
        if count > max_count:
            max_count = count
            most_freq = i
    return most_freq
t = (1, 2, 2, 3, 4, 2, 5, 1)
print(frequent(t))
```

```python
'''9.Write a python function to Given a tuple of integers, return a tuple of all the digits used.
'''
def tup(t):
    digits = []
    for i in t:
        for d in str(i):
            if d != '-':           
                digits.append(int(d))
    return tuple(digits)
t = (12, -345, 67)
result = tup(t)
print(result)
```

```python
'''10.Write a python function to Return a tuple of strings, each reversed individually.'''
def rev(t):
    return tuple(s[::-1] for s in t)
t = ("hello", "world", "python")
reversed_t = rev(t)
print(reversed_t)
```
