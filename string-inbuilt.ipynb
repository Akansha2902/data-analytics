```python
'''13. Find the longest palindromic substring in a given string.'''
s= input("Enter a string:-")
long= ""
for i in range(len(s)):
    for j in range(i, len(s)):
        sub = s[i:j+1]
        if sub == sub[::-1]:     # check palindrome
            if len(sub) > len(long):
                long= sub
print(f"Longest palindrome:{long}")
```

```python
'''14. Compute the longest common prefix and suffix of two strings.'''
n=input("enter a string:-")
m=input("enter a string:-")
prefix=""
l=min(len(n),len(m))
for i in range(l):
    if n[i]==m[i]:
        prefix+=n[i]
    else:
        break 
print(prefix)
n1=len(n)-1
m1=len(m)-1
suffix=""
while n1>=0 and m1>=0 and n[n1]==m[m1]:
    suffix=n[n1]+suffix
    n1-=1
    m1-=1
print(suffix)
```

```python
'''21. Count word frequencies in a paragraph and print them in descending order.'''
text = input("Enter a paragraph:-").lower()
words = text.split()
unique= []
counts = []
for w in words:
    if w in unique:
        counts[unique.index(w)] += 1
    else:
        unique+=w
        counts+=1
for i in range(len(counts)):
    for j in range(i+1, len(counts)):
        if counts[i] < counts[j]:
            counts[i], counts[j] = counts[j], counts[i]
            unique[i], unique[j] = unique[j], unique[i]
for i in range(len(unique)):
    print(unique[i], ":", counts[i])
```

```python
'''25. Generate all unique subsets of characters in lexicographical order.'''
s = input("Enter a string: ")  
s = ''.join(sorted(s))          
subsets = set()
n = len(s)
for mask in range(1 << n):      
    subset = ""
    for i in range(n):
        if mask & (1 << i):
            subset += s[i]
    subsets.add(subset)
result = sorted(subsets)
for sub in result:
    print(sub)
```

```python
'''26. Find the longest substring with no repeating characters.'''
s = input("Enter a string:-")
c= ""      
longest = ""      
for i in s:
    if i in c:
        idx = c.index(i)
        c= c[idx+1:]   
    c+= i                 
    if len(c) > len(longest):
        longest = c
print(longest)
```
