```python
'''3. Write a Python program that defines a user-defined function get_line_equation(x1, y1, x2, y2) which takes the coordinates of two points (x1, y1)
and (x2, y2) and returns the slope and y-intercept of the line passing through them. The function should return a tuple (m, c) representing the slope (m)
 and y-intercept (c) in the slope-intercept form (y = mx + c). Handle the case of vertical lines (undefined slope) appropriately (e.g., return None for slope
 and the x-intercept).
'''
def get_line_equation(x1, y1, x2, y2):
    if x1 == x2:
        return None, x1
    else:
        m = (y2 - y1) / (x2 - x1)
        c = y1 - m * x1
        return m, c

print(get_line_equation(0, 0, 2, 2))
print(get_line_equation(1, 2, 1, 5))
```

```python
'''6. Create a Python program with a user-defined function calculate_factorial(n) that takes a non-negative integer n as input and returns its factorial.
The factorial of a number n is the product of all positive integers less than or equal to n. For example, 5 factorial (5!) is 5 * 4 * 3 * 2 * 1 = 120. '''
def calculate_factorial(n):
    if n < 0:
        return None
    factorial = 1
    for i in range(1, n + 1):
        factorial *= i
    return factorial

num = int(input("Enter a number: "))

result = calculate_factorial(num)

if result is None:
    print(" number is negative.")
else:
    print(f"The factorial of {num} is {result}")

```

```python
'''7. Write a Python program that defines a user-defined function calculate_mean(numbers_list) which takes a list of numbers as input and returns their
arithmetic mean (average). The mean is the sum of all numbers divided by the count of numbers.
'''
def calculate_mean(numbers_list):
    if len(numbers_list) == 0:
        return None
    total = 0
    for num in numbers_list:
        total += num
    return total / len(numbers_list)

nums = input("Enter numbers separated by spaces: ").split()
nums = [int(n) for n in nums]
print(calculate_mean(nums))


```

```python
'''8. Write a Python program that defines a user-defined function sum_list_elements(numbers_list) which takes a list of numbers as input and returns the
sum of all elements in the list. You should not use the built-in sum() function. Use a loop instead.
'''
def sum_list_elements(numbers_list):
    if len(numbers_list) == 0:
        return None
    total = 0
    for num in numbers_list:
        total += num
    return total

nums = input("Enter numbers separated by spaces: ").split()
nums = [int(n) for n in nums]
print(sum_list_elements(nums))
```

```python
'''9. Create a Python program with a user-defined function find_largest_element(numbers_list) that takes a list of numbers as input and returns the largest
 element in that list. You should not use the built-in max() function. Use a loop and a variable to keep track of the largest number found so far.
'''
def find_largest_element(numbers_list):
    if len(numbers_list) == 0:
        return None
    largest = numbers_list[0]
    for num in numbers_list:
        if num > largest:
            largest = num
    return largest

nums = input("Enter numbers separated by spaces: ").split()
nums = [int(n) for n in nums]
print(find_largest_element(nums))
```

```python
'''1. Write a python function to find  Greatest Common Divisor (GCD) using Euclid’s algorithm'''
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return a

x = int(input("Enter first number: "))
y = int(input("Enter second number: "))

print(f"GCD of {x} and {y} is {gcd(x, y)}")
```

```python
'''3.Write a python function to find Equation of the perpendicular bisector of segment (x1, y1)-(x2, y2)'''
def perpendicular_bisector(x1, y1, x2, y2):
    mid_x = (x1 + x2) / 2
    mid_y = (y1 + y2) / 2
    if x1 == x2:
        return 0, mid_y
    if y1 == y2:
        return None, mid_x
    m = (y2 - y1) / (x2 - x1)
    m_perp = -1 / m
    c = mid_y - m_perp * mid_x
    return m_perp, c

x1, y1 = 2, 3
x2, y2 = 6, 7

print(perpendicular_bisector(x1, y1, x2, y2))
```

```python
'''4.Write a python function to find Area and perimeter of a triangle using coordinates'''
import math

def triangle_area_perimeter(x1, y1, x2, y2, x3, y3):
    area = 0.5 * abs(x1*(y2 - y3) + x2*(y3 - y1) + x3*(y1 - y2))
    side1 = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
    side2 = math.sqrt((x3 - x2)**2 + (y3 - y2)**2)
    side3 = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)
    perimeter = side1 + side2 + side3
    return area, perimeter
x1, y1 = 0, 0
x2, y2 = 4, 0
x3, y3 = 0, 3
area, perimeter = triangle_area_perimeter(x1, y1, x2, y2, x3, y3)
print(f"Area of triangle: {area}")
print(f"Perimeter of triangle: {perimeter}")
```

```python
'''5.Write a python function to Check if point (px, py) lies inside/on/outside a circle with center (cx, cy) and radius r'''
def point_circle_relation(px, py, cx, cy, r, eps=1e-9):
    if r < 0:
        raise ValueError("Radius must be non-negative")
    dist_sq = (px - cx) ** 2 + (py - cy) ** 2
    r_sq = r ** 2
    if dist_sq < r_sq - eps:
        return "inside"
    elif abs(dist_sq - r_sq) <= eps:
        return "on"
    else:
        return "outside"
print(point_circle_relation(1, 1, 0, 0, 2))
print(point_circle_relation(2, 0, 0, 0, 2))
print(point_circle_relation(3, 0, 0, 0, 2))
```
