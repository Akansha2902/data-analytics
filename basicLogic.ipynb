```python
''' 1.Write a python program to Find the maximum number in a list'''
n=[34,2,3,12,32,67,89,81,99]
max=n[0]
for i in n:
  if i>max:
    max=i
print(max)
```

```python
''' 2. Write a python program to Count how many strings contain the letter ‘a' in a list of strings'''
n=['kjfiuas','akansha','ahauibuaiua','udicba']

for i in n:
  count=0
  for j in i:
    if j=='a':
      count+=1
  print(f"{i} having a ={count}")
```

```python
'''3. Write a python program to Find numbers which are squares of other numbers from a list of numbers '''
n=[2,3,4,8,9,16,5]
square =[]
for i in n:
  for sq in n:
    if i==sq**2:
      square+=[i]
print(square)
```

```python
''' 4. Write a python program to Find all prime numbers from a list'''
n = [2, 3, 4, 5, 10, 13, 17, 18, 19, 20, 23, 29, 30]
prime = []
for i in n:
    if i > 1:
        count = 0
        for j in range(1, i + 1):
            if i % j == 0:
                count += 1
        if count == 2:
            prime+=[i]
print(prime)
```

```python
''' 5. Write a python program to Count vowels in each string of a list'''#a, e, i, o, or u
n=['orange','jdsbins','dviwusb','opwidu','iuqwoua']
for i in n:
  count=0
  for j in i:
    if j in ('a','i','o','e','u','A','E','I','O','U'):
      count+=1
  print(f"{i} having {count} vowels")
```

```python
'''7.Write a python program to Find the average of even numbers only from a list'''
n=[43,12,24,4,89,90,45,64]
count=0
total=0
for i in n:
  if i%2==0:
    count+=1
    total+=i
avg=total/count
print(avg)
```

```python
''' 8.Write a python program to Find the product of even numbers and reduction of odd numbers in a list'''
n=[4,23,12,14,8,95,7,2,6]
product=1
reduction =1
for i in n:
  if i%2==0:
    product*=i
  else:
    reduction-=i
print(f"Product = {product}")
print(f"Reduction = {reduction}")
```

```python
''' 9.Write a python program to Merge two lists without duplicates'''
list1 = [1, 2, 3, 4, 5]
list2 = [4, 5, 6, 7, 8]
all= list1 + list2
merged = []
for i in all:
    for x in merged:
        if x == i:
            break
    else:
        merged+=[i]
print(merged)
```

```python
''' 10. Write a python program to Find the difference between the largest and smallest number in a list'''
n=[34,2,3,12,32,67,89,81,99]
max=n[0]
for i in n:
  if i>max:
    max=i
print(f"maximum = {max}")
min=n[0]
for j in n:
  if j<min:
    min=j
print(f"minimum = {min}")
diff=max-min
print(f"difference = {diff}")
```
