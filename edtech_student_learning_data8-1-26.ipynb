```python
#35. Create a new column attendance_flag based on attendance percentage
def attendance_flag(x):
    if x>=75:
        return 'green flag'
    else:
        return "Red flag"

df["attendance_flag"]=df["attendance_pct"].apply(attendance_flag)
```

```python
df.head()
```

```python
#36. Create a new column learning_intensity based on learning hours
def learning_intensity(x):
    if x>40:
        return"High"
    elif x>=20:
        return"medium"
    else:
        return"low"
df["learning_intensity"]=df["learning_hours"].apply(learning_intensity)
    
```

```python
df.head()
```

```python
#37. Classify students based on total_score using apply().
def classify_score(x):
    if x>=200:
        return"Excellent"
    elif x>=150:
        return"Average"
    else:
        return"Poor"
df["Classify_students"]=df["total_score"].apply(classify_score)
```

```python
df.head()
```

```python
#38. Create a placement eligibility column using apply().
def placement_eligibility(x):
    if x["total_score"]>=150 and x["attendance_pct"]>=75:
        return"Eligible"
    else:
        return"Not Eligible"
df["placement_eligibility"]=df.apply(placement_eligibility,axis=1)
```

```python
df.head()
```

```python
#39. Map placement_status values to numerical values.
num={"Placed":1,"Not Placed":0}
df["Num_Placement_Status"]=df["placement_status"].map(num)
```

```python
#40. Map batch_type values to abbreviated codes.
code={"Weekday":"WD","Weekend":"WE"}
df["Batch_type_code"]=df["batch_type"].map(code)
```

```python
#41. Map gender values to single-letter codes.
g={"Male":"M","Female":"F"}
df["Gender_code"]=df["gender"].map(g)
```
