```python
import pandas as pd
```

```python
df=pd.read_csv("Downloads/edtech_student_learning_data.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
#1. Group the data by course and calculate the average attendance percentage.
df.groupby("course")["attendance_pct"].mean()
```

```python
#2. Group the data by city and count the number of students in each city.
df.groupby("city")["student_id"].count()
```

```python
#3. Group the data by batch_type and calculate the average quiz score.
df.groupby("batch_type")["avg_quiz_score"].mean()
```

```python
#4. Group the data by placement_status and calculate the average total_score.
df["total_score"]=df["avg_quiz_score"]+df["assignment_score"]+df["project_score"]
df.groupby("placement_status")["total_score"].mean()
```

```python
#5. Group the data by course and city and calculate the number of students in each combination
df.groupby(["course","city"])["student_id"].count()
```

```python
#6. Group the data by course and calculate the maximum, minimum, and average project score.
df.groupby("course")["project_score"].agg(["max","min","mean"])
```

```python
#7. Group the data by city and calculate total and average learning hours.
df.groupby("city")["learning_hours"].agg(["sum","mean"])
```

```python
#8. Group the data by course and placement_status and count students.
df.groupby(["course","placement_status"])["student_id"].count()
```

```python
#9. Filter grouped results to display only courses with average attendance percentage greater than 80.
avg_attendance_pct=df.groupby("course")["attendance_pct"].mean()
avg_attendance_pct.head()
```

```python
avg_attendance_pct[avg_attendance_pct > 80]
```

```python
#10. Filter grouped results to show only cities with more than 50 students.
no_student=df.groupby("city")["student_id"].count()
no_student.head()
```

```python
no_student[no_student >50]
```

```python
#11. Perform multiple aggregations on course to calculate average quiz score, average assignment score, and average project score.
df.groupby("course").agg(avg_quiz=("avg_quiz_score","mean"),
                         avg_assign_score=("assignment_score","mean"),
                         avg_project_score=("project_score","mean"))
```

```python
#12. Perform named aggregations on course to calculate average_attendance, max_project_score, and mean_total_score.
df.groupby("course").agg(average_attendance=("attendance_pct","mean"),
                         max_project_score=("project_score","max"),
                         mean_total_score=("total_score","mean"))
```

```python
#13. Perform named aggregations on city to calculate total_students, placed_students, and average_salary.
df["place"]=df["placement_status"]=="Placed"
df.groupby("city").agg(total_students=("student_id","count"),
                       placed_students=("place","sum"),
                       average_salary=("salary_offered","mean"))
```
