```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

**Create a list of 7 flowers and perform label encoding on them**

```python
flowers =["rose","marigold","Lilly","rose","Lilly","sunflower","marigold","rose"]
encoder=LabelEncoder()
flowers_encoded=encoder.fit_transform(flowers)
print(flowers_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

**create a dataframe with 5 columns(categorical+numerical)and perform one hot encoding**

```python
data = {
    "gender":["male","male","female","male","female","female"],
    "department":["IT","sale","sale","IT","HR","HR"],
    "city":["mumbai","aurangabad","pune","pune","mumbai","pune"],
    "salary":[10000,20000,13000,45000,10000,50000],
    "age":[23,45,19,34,26,25]
}
df = pd.DataFrame(data)
print(df)
```

```python
df=pd.DataFrame({"Dep":df["department"]})
encoded_df=pd.get_dummies(df,columns=["Dep"],dtype=int)
print(encoded_df)
```

**create a list of ordinal data with 6 different ordinal values and perform ordinal encoding on them**

```python
from sklearn.preprocessing import OrdinalEncoder
dressSize = [["XL"],["S"],["M"],["S"],["XL"],["M"],["XS"],["XXL"],["S"],["XXS"]]
encoder = OrdinalEncoder(categories=[["XXS","XS","S","M","XL","XXL"]])
encoded_size = encoder.fit_transform(dressSize)
print(encoded_size)
```
