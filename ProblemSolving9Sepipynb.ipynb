```python
'''1.Given a time in 24‑hour format as a four‑digit integer or string (e.g., "1425", 1425, "0907"), convert it to spoken English using a 12‑hour clock'''
def time(time_str):
    time_str = str(time_str)
    while len(time_str) < 4:
        time_str = "0" + time_str

    hour = int(time_str[0]) * 10 + int(time_str[1])
    minute = int(time_str[2]) * 10 + int(time_str[3])

    if hour == 0:
        s = 12
        m = "AM"
    elif hour < 12:
        s = hour
        m = "AM"
    elif hour == 12:
        s = 12
        m = "PM"
    else:
        s = hour - 12
        m = "PM"

    if minute == 0:
        return str(s) + " o'clock " + m
    elif minute < 10:
        return str(s) + " oh " + str(minute) + " " + m
    else:
        return str(s) + " " + str(minute) + " " + m

print(time("1425"))
print(time(907))
print(time("0000"))
```

```python
'''2.Write a program to find how many distinct pairs (a, b) of non-negative integers satisfy the equation a² + b² = n.'''
def find_pairs(n):
    pairs = []
    a = 0
    while a * a <= n:
        b_squared = n - a * a
        b = 0
        while b * b <= b_squared:
            if a * a + b * b == n:
                found = False
                for (x, y) in pairs:
                    if x == b and y == a:
                        found = True
                        break
                if not found:
                    pairs.append((a, b))
            b += 1
        a += 1
    return pairs

n = 25
pairs = find_pairs(n)
print("Pairs:", pairs)
print("Number of pairs:", len(pairs))
```

```python
'''3.Write a program that reads an integer and returns the sum and the product of its first and last digits.'''
def sum_and_product(n):
    if n < 0:
        n = -n

    num_str = str(n)
    first_digit = int(num_str[0])
    last_digit = int(num_str[len(num_str) - 1])

    digit_sum = first_digit + last_digit
    digit_product = first_digit * last_digit
    return digit_sum, digit_product

num = 5832
s, p = sum_and_product(num)
print("Sum:", s, "Product:", p)
```

```python
'''4.Write a program to calculate the Pearson correlation coefficient for two lists of equal length.'''
def pearson_correlation(x, y):
    if len(x) != len(y):
        return("Lists must have equal length")

    n = len(x)
    total_x = 0
    total_y = 0
    i = 0
    while i < n:
        total_x += x[i]
        total_y += y[i]
        i += 1
    mean_x = total_x / n
    mean_y = total_y / n

    numerator = 0
    denominator_x = 0
    denominator_y = 0

    i = 0
    while i < n:
        diff_x = x[i] - mean_x
        diff_y = y[i] - mean_y
        numerator += diff_x * diff_y
        denominator_x += diff_x * diff_x
        denominator_y += diff_y * diff_y
        i += 1

    denominator = (denominator_x * denominator_y) ** 0.5
    if denominator == 0:
        return 0
    return numerator / denominator

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]
pearson_correlation(x, y)
```

```python
'''5.Write a program to calculate the mean absolute deviation of a list of numbers.'''
def mean_absolute_deviation(d):
    n = len(d)
    total = 0
    i = 0
    while i < n:
        total += d[i]
        i += 1
    mean = total / n

    total_deviation = 0
    i = 0
    while i < n:
        deviation = d[i] - mean
        if deviation < 0:
            deviation = -deviation
        total_deviation += deviation
        i += 1

    return total_deviation / n

d = [1, 2, 3, 4, 5]
mean_absolute_deviation(d)
```

```python
'''6.Write a program that calculates the moving average of a list of numbers given a window size.'''
def moving_average(d, window_size):
    if window_size <= 0:
        return("Window size must be positive")
    if window_size > len(d):
        return("Window size cannot be greater than data length")

    averages = []
    i = 0
    while i <= len(d) - window_size:
        total = 0
        j = i
        while j < i + window_size:
            total += d[j]
            j += 1
        avg = total / window_size
        averages.append(avg)
        i += 1

    return averages

d = [1, 2, 3, 4, 5, 6]
window_size = 3
moving_average(d, window_size)
```

```python
'''1.Create a function-based program that calculates the lower and upper quartiles of a dataset. The dataset should be processed without using built-in
 functions or sorting utilities. Implement your own logic to divide the data into appropriate segments and extract quartile values.'''
def sort_list(data):
    for i in range(len(data)):
        for j in range(i+1, len(data)):
            if data[j] < data[i]:
                temp = data[i]
                data[i] = data[j]
                data[j] = temp
    return data

def median(data):
    n = len(data)
    if n % 2 == 0:
        return (data[n//2 - 1] + data[n//2]) / 2
    else:
        return data[n//2]

def quartiles(data):
    data = sort_list(data)
    n = len(data)
    if n % 2 == 0:
        lower_half = data[:n//2]
        upper_half = data[n//2:]
    else:
        lower_half = data[:n//2]
        upper_half = data[n//2+1:]
    q1 = median(lower_half)
    q3 = median(upper_half)
    return q1, q3

print(quartiles([7,1,3,5,9,11,13]))
```

```python
'''2.Build a program using only user-defined functions to determine whether a sequence of numbers is increasing, decreasing, or neither. The solution
must not rely on any built-in comparison helpers or sequence manipulation functions.'''
def type(seq):
    increasing = True
    decreasing = True
    for i in range(1, len(seq)):
        if seq[i] > seq[i-1]:
            decreasing = False
        elif seq[i] < seq[i-1]:
            increasing = False
        else:
            increasing = False
            decreasing = False
    if increasing:
        return "Increasing"
    elif decreasing:
        return "Decreasing"
    else:
        return "Neither"

print(type([1,2,3,4]))
print(type([5,4,3,2,1,0]))
print(type([23,45,12,3,1]))
```

```python
'''3. Write a program to solve a system of two linear equations with two variables. Construct all logic from scratch, using only user-defined functions
 for calculations and comparisons. Your solution should handle different types of coefficients and ensure a unique solution is found when possible.'''
def linear(a, b, c, d, e, f):
    det = a*e - b*d
    if det == 0:
        return "No unique solution"
    x = (c*e - b*f) / det
    y = (a*f - c*d) / det
    return x, y

print(linear(2,1,5,1,-1,1))
```

```python
'''4.Construct a program that evaluates a polynomial expression of a single variable for a given input value. For example, evaluate an expression
like ax² + bx + c at x = n using only user-defined arithmetic functions.'''
def multi(a, b):
    result = 0
    for i in range(abs(b)):
        result += a
    if b < 0:
        result = -result
    return result

def power(x, n):
    result = 1
    for i in range(n):
        result = multi(result, x)
    return result

def poly(a, b, c, x):
    return multi(a, power(x, 2)) + multi(b, x) + c

print(poly(2,3,4,2))
```

```python
'''5.Write a program using user-defined functions that computes the weighted average of a set of values with associated weights.
Implement all operations manually, including multiplication, summation, and division. Ensure the logic checks for mismatched lengths between values
and weights.'''
def multiply(a, b):
    result = 0
    for i in range(abs(b)):
        result += a
    if b < 0:
        result = -result
    return result
def average(values, weights):
    if len(values) != len(weights):
        return "Error: mismatched lengths"
    total_weighted = 0
    total_weight = 0
    for i in range(len(values)):
        total_weighted += multiply(values[i], weights[i])
        total_weight += weights[i]
    return total_weighted / total_weight

print(average([3,5,7],[2,4,6]))
```

```python
'''6.Design a program that identifies whether a given list contains any significant outliers. Use custom logic to assess how far individual values
deviate from the main body of data. All statistical measures used must be computed using your own user-defined functions.'''
def quartiles(data):
    data = sort_list(data)
    n = len(data)
    if n % 2 == 0:
        lower_half = data[:n//2]
        upper_half = data[n//2:]
    else:
        lower_half = data[:n//2]
        upper_half = data[n//2+1:]
    q1 = median(lower_half)
    q3 = median(upper_half)
    return q1, q3
def outliers(data):
    q1, q3 = quartiles(data)
    iqr = q3 - q1
    lower = q1 - 1.5 * iqr
    upper = q3 + 1.5 * iqr
    outlier = []
    for v in data:
        if v < lower or v > upper:
            outlier.append(v)
    return outlier

print(outliers([1,2,2,2,3,100]))
```

```python
'''7.Construct a program to analyze two numerical datasets and determine the strength and direction of their linear relationship. Use only user-defined
 functions to perform all required calculations, avoiding any shortcuts or statistical libraries.'''
def mean(data):
    total = 0
    for v in data:
        total += v
    return total / len(data)

def correlation(x, y):
    n = len(x)
    x_mean = mean(x)
    y_mean = mean(y)
    num = 0
    den_x = 0
    den_y = 0
    for i in range(n):
        dx = x[i] - x_mean
        dy = y[i] - y_mean
        num += dx * dy
        den_x += dx * dx
        den_y += dy * dy
    return num / ((den_x**0.5) * (den_y**0.5))

print(correlation([1,2,3],[2,4,6]))
```

```python
'''8.Create a program using user-defined functions that takes a linear algebraic expression with like terms (e.g., 2x + 3x - x + 5) and
 simplifies it to its shortest form. Ensure that your function handles sign changes and coefficient aggregation manually.'''
def expression(terms):
    coeff = 0
    constant = 0
    for t in terms:
        if "x" in t:
            if t == "x":
                coeff += 1
            elif t == "-x":
                coeff -= 1
            else:
                coeff += int(t.replace("x",""))
        else:
            constant += int(t)
    result = ""
    if coeff != 0:
        result += str(coeff) + "x"
    if constant != 0:
        if result != "" and constant > 0:
            result += "+"
        result += str(constant)
    return result

print(expression(["2x","+3x","-x","+5"]))
```

```python
'''9.Write a set of user-defined functions to check whether a given sequence of numbers forms an arithmetic progression. If so, determine the
common difference and the general term. All computations must be derived using basic arithmetic logic.'''
def AP(seq):
    if len(seq) < 2:
        return "Not enough terms"
    d = seq[1] - seq[0]
    for i in range(2, len(seq)):
        if seq[i] - seq[i-1] != d:
            return "Not AP"
    return "AP with difference " + str(d)

print(AP([2,4,6,8,10]))
```

```python
'''10.Develop a function-based solution that determines the most frequently occurring value in a dataset. Implement custom logic to count occurrences
and identify the maximum frequency. Do not use dictionaries, built-ins, or collections.'''
def mode(data):
    max_count = 0
    mode_value = None
    for i in range(len(data)):
        count = 0
        for j in range(len(data)):
            if data[i] == data[j]:
                count += 1
        if count > max_count:
            max_count = count
            mode_value = data[i]
    return mode_value

print(mode([1,2,2,3,3,3,4]))
```
