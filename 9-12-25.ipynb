```python
'''1. Write a Python program to find the unique employee IDs in a company.'''
try: 
    employee_ids = [101, 102, 103, 101, 104, 105, 103, 106, 102]
    u=[]
    for i in employee_ids:
        if employee_ids.count(i)==1:
            u+=[i]
except e:
    print("an error occurs while finding a unique employee ids")
finally:
    print(u)
```
