```python
'''Take a string and a character from the user. Count how many times the character appears in the string.'''
n=input("enter a string:-")
m=input("enter a character:-")
count=0
for i in n:
    if m == i:
      count+=1
      
print(f"in {n} string the character {m} is appears {count} times.")
 
```

```python
'''Take a sentence, a word to replace, and the new word from the user. Replace it in the sentence.'''
sentence = input("Enter a sentence: ")
old_word = input("Enter the word to replace: ")
new_word = input("Enter the new word: ")
new_sentence =""
for i in sentence:
    if i==old_word:
        new_sentence +=new_word
    else:
        new_sentence+=i
print(new_sentence)
```

```python
'''7.Take a comma-separated string from the user and split it into a list.'''
s = input("Enter a comma-separated string: ")
result = []
c = ""
for ch in s:
    if ch == ",":
        result+=[c]
        c = ""
    else:
        c += ch
result+=[c]
print("List:", result)
```

```python
'''Take a string input and check whether it is a palindrome (same forward and backward).'''
s = input("Enter a string: ")
palindrome = True
count=0
for i in s:
    count+=1
left = 0
right = count - 1
while left < right:
    if s[left] != s[right]:
        palindrome = False
        break
    left += 1
    right -= 1
if palindrome==True:
    print("It is a palindrome.")
else:
    print("It is not a palindrome.")
```

```python
'''Take an email address from the user and extract the domain name (text after the @).'''
email = input("Enter an email address: ")
domain = ""
found = False
for ch in email:
    if found:
        domain += ch
    if ch == "@":
        found= True
print("Domain name:", domain)
```
