```python
import pandas as pd
data = {
    "A":[12,24,76889,34,234,1243],
    "B":[45,56,657,356,2135,2345],
    "c":[56658,87098,2134,234,000,0],
    "d":[00,234,0,234,134,0]
}
```

```python
df=pd.DataFrame(data)
```

```python
df
```

**MIN-MAX SCALER**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_minmax = df.copy()
df_minmax = pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
df_minmax
```

**Z-SCORE NORMALIZATION**

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_standard = df.copy()
df_standard = pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
df_standard
```

**ROBUST-SCALER**

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
df_robust = df.copy()
df_robust = pd.DataFrame(
    scaler.fit_transform(df),
    columns=df.columns
)
df_robust
```
