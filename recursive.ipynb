```python
'''1. Write a recursive function to return the sum of first n natural numbers
ex: sum(5) = 5 + 4 + 3 + 2 + 1
Take the n (number) from user'''
def sum(n):
    if n == 0:
        return 0
    else:
        return n + sum(n - 1)

num = int(input("Enter a number: "))
print(f"Sum of 1st {num} natural numbers is:",sum(num))
```

```python
'''2. Write a recursive function to calculate a ^ b without using exponent operator '''
def power(a, b):
    if b == 0:
        return 1
    else:
        return a * power(a, b - 1)

a = int(input("Enter base (a): "))
b = int(input("Enter exponent (b): "))
print(f"{a}^{b} =", power(a, b))
```

```python
'''3. Write a recursive function to count number of digits in the number entered by the user:
ex: if user enterd = 1234 - The count of the digit is 4'''
def count(n):
    if n == 0:
        return 0
    else:
        return 1 + count(n // 10)

num = int(input("Enter a number: "))
if num == 0:
    print("Number of digits in 0 is: 1")
else:
    print(f"Number of digits in {num} is:", count(num))
```
