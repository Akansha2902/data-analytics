```python
'''Write a python program to Replace all vowels in strings with their count in that word'''
s= input("Enter a sentence: ")
vowels = "aeiouAEIOU"
final= ""
words = s.split()
for word in words:
    count = 0
    for ch in word:
        if ch in vowels:
            count += 1
    new= ""
    for ch in word:
        if ch in vowels:
            new+= str(count)
        else:
            new+= ch
    final+= new+ " "
print(final)
```

```python
'''4. Write a python program to Keep only strings where each character appears exactly once'''
s=['cat','hello','python','puppy']
result=[]
for i in s:
    seen=""
    repeat=False
    for n in i:
        if n in seen:
            repeat=True
            break
        seen+=n
    if not repeat:
        result+=[i]
print(result)
```

```python
'''2. Write a python program using inbuilt function to Filter strings that alternate vowels and consonants'''
a = ['apple','dog','cat','puppy','pinapple']   
vowels = "aeiouAEIOU"
result = []
for word in a:
    ok = True
    count = 0
    for ch in word:
        count += 1
    for i in range(count - 1):
        c1 = word[i]
        c2 = word[i + 1]
        is_v1 = c1 in vowels
        is_v2 = c2 in vowels
        if is_v1 == is_v2:
            ok = False
            break
    if ok:
        result+=[word]
print(result)
```

```python
'''3. Write a python program using inbuilt function to Keep strings that contain at least 3 different vowels'''
a = ['apple','dog','cat','puppy','pinapple']   
result=[]
for word in a:
    v=w=x=y=z=0
    for ch in word:
        if ch=='a' or ch=='A':
            v=1
        elif ch=='e' or ch=='E':
            w=1
        elif ch=='o' or ch=='O':
            x=1
        elif ch=='i' or ch=='I':
            y=1
        elif ch=='u' or ch=='U':
            z=1
    count=0
    count+=v
    count+=w
    count+=x
    count+=y
    count+=z
    if count>=3:
        result+=[word]
print(result)
```

```python
'''7.From a sentence, return all words that appear more than once'''
sentence = "apple dog cat apple puppy dog pinapple dog pinapple"
words = []
current = ""
for ch in sentence:
    if ch != " ":            
        current += ch
    else:
        if current != "":    
            words+=[current]
            current = ""
if current != "":
    words+=[current]
freq = {}
for w in words:
    if w in freq:
        freq[w] += 1
    else:
        freq[w] = 1
result = []
for w in freq:
    if freq[w] > 1:
        result+=[w]
print(result)
```

```python
'''8. Write a python program using inbuilt function to Return list of (index, word) for words that are palindromes'''
words = ['level', 'dog', 'madam', 'cat', 'noon', 'apple']
result = []
i = 0     
for word in words:
    if word == word[::-1]:     
        result+=[(i, word)]
    i += 1                      
print(result)
```

```python
'''10. Write a python program using inbuilt function to Filter strings that match their sorted version (alphabetical)'''
words = ["abc", "dog", "aet", "almost", "xy", "caa"]
result = []
for w in words:
    ok = True
    first = True
    prev = ""
    for ch in w:         
        if first:
            prev = ch    
            first = False
        else:
            if prev > ch:    
                ok = False
            prev = ch       
    if ok:
        result+=[w]
print(result)
```
