```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("edtech_student.csv")
```

```python
df.head()
```

## Label encoding

```python
G=df["gender"]
encoder=LabelEncoder()
G_encoded=encoder.fit_transform(G)
print(G_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
C=df["city"]
encoder=LabelEncoder()
C_encoded=encoder.fit_transform(C)
print(C_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
Course=df["course"]
encoder=LabelEncoder()
Course_encoded=encoder.fit_transform(Course)
print(Course_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
batch=df["batch_type"]
encoder=LabelEncoder()
batch_encoded=encoder.fit_transform(batch)
print(batch_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
placement =df["placement_status"]
encoder=LabelEncoder()
placement_encoded=encoder.fit_transform(placement)
print(placement_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

## one hot encoding

```python
df=pd.DataFrame({"gen":df["gender"]})
encoded_df=pd.get_dummies(df,columns=["gen"],dtype=int)
print(encoded_df)
```
