```python
'''4. Write a python function to find the total number of vowels in a string '''
def count_vowels(n):
    vowels = "aeiouAEIOU"
    count = 0
    for char in n:
        if char in vowels:
            count += 1
    return count

string = input("Enter a string: ")
print("Total number of vowels:", count_vowels(string))
```

```python
'''5. Write a python function to calculate distance between two points'''
def distance(x1, y1, x2, y2):
    dist = ((x2 - x1) ** 2 + (y2 - y1) ** 2) ** 0.5
    return dist

x1 = int(input("Enter x1: "))
y1 = int(input("Enter y1: "))
x2 = int(input("Enter x2: "))
y2 = int(input("Enter y2: "))

print("Distance between points:", distance(x1, y1, x2, y2))
```

```python
'''3.Write a Python function to check whether a number falls within a given range.'''
def num_range():
  n=int(input("enter a number:"))
  start=int(input("enter a number to start a range:"))
  end=int(input("enter a number to end the range:"))
  if start<=n <=end:
    return f"the {n} is in the range {start} to {end}"
  else:
    return f"the {n} is not in the range"

print(num_range())
```

```python
'''6. Write a Python program with a user-defined function is_right_triangle(side1, side2, side3) that takes the lengths of three sides of a triangle as input
and returns True if it can form a right-angled triangle (using the Pythagorean theorem), and False otherwise. '''

def is_right_triangle(side1, side2, side3):
    a, b, c = side1,side2,side3
    return c**2 == a**2 + b**2

s1 = int(input("Enter side 1: "))
s2 = int(input("Enter side 2: "))
s3 = int(input("Enter side 3: "))

if is_right_triangle(s1, s2, s3):
    print("Yes, it is a right-angled triangle.")
else:
    print("No, it is not a right-angled triangle.")
```
