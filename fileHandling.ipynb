```python
'''8. Write a program to count vowels and consonants in a file.'''
with open("info.txt",'r') as file:
    c=file.read()
    vowel='aeiouAEIOU'
    v_count=0
    c_count=0
    for i in c:
        if i.isalpha():
            if i in vowel:
                v_count+=1
            else:
                c_count+=1
    print(f"count of vowel={v_count} and consonants={c_count}")
    
```

```python
'''14. Write a program to count frequency of each word in a file.'''
with open("info.txt",'r') as file:
    c=file.read()
    s=c.split()
    dup=[]
    for i in s:
        if i not in dup:
            dup+=[i]
    print(dup)
    for n in dup:
        print(f"{n} count={s.count(n)}")
```

```python
'''18. Write a program to read numbers from a file and find their sum & average.'''
with open("info.txt",'r') as file:
    c=file.read()
    s=c.split()
    d=[]
    for i in s:
        for c in i:
            if c.isdigit():
                d+=[c]
    print(d)
    u=len(d)
    add=0
    for j in d:
        add+=int(j)
    print("sum of all digits=",add)
    avg=add/u
    print("average of all digits=",avg)
    
                
```

```python
'''20. Write a program that reads a file and prints number of uppercase, lowercase, digits, and special symbols.'''
with open("info1.txt",'r')as file:
    c=file.read()
    u=[]
    l=[]
    d=[]
    s=[]
    sp=c.split()
    for i in sp:
        for j in i:
            if j.isupper():
                u+=[j]
            if j.islower():
                l+=[j]
            if j.isdigit():
                d+=[j]
            if j in "!@#$%^&*()-_=+[]{}|;:'\",.<>?/\\`~":
                s+=[j]
    print(f"number of uppercase={len(u)} lowercase={len(l)} digit={len(d)} special symbols={len(s)}")
    
    
```
