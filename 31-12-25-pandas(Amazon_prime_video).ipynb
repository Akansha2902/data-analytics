```python
#29. Create a new column popularity_score as views_millions + likes_millions.
df["popularity_score"]=df["views_millions"]+df["likes_millions"]
df.head()
```

```python
#30. Create a new column rating_category using apply().
def example(x):
    if x >=8:
        return "High"
    elif x>=5:
        return "Medium"
    else:
        return "Low"
df["rating_category"]=df["imdb_rating"].apply(example)
df.head()
```

```python
#31. Map subscription_plan to short codes.
code={"Basic":"B","Standard":"S","Mobile":"M","Premium":"P"}
df["subscription_plan_code"]=df["subscription_plan"].map(code)
```

```python
#32. Map is_trending values to numerical format.
code={"No":0,"Yes":1}
df["is_trending_code"]=df["is_trending"].map(code)
```
