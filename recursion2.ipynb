```python
'''1. Write a python program to Count Even Digits in a Number using recursion '''
def count_even_digits(n):
    if n == 0:
        return 0
    else:
        digit = n % 10
        if digit % 2 == 0:
            return 1 + count_even_digits(n // 10)
        else:
            return count_even_digits(n // 10)

num = 243680
count_even_digits(num)
```

```python
'''2. Write a python program to Replace All Occurrences of 0 with 5 using recursion'''
def replace(n):
    if n == 0:
        return 5
    elif n < 10:
        if n == 0:
            return 5
        else:
            return n
    else:
        digit = n % 10
        if digit == 0:
            digit = 5
        return replace(n // 10) * 10 + digit

num = 102030
replace(num)
```

```python
'''3. Write a python program to Find Number of Zeroes in a Number using recursion'''
def count(n):
    if n == 0:
        return 1
    elif n < 10:
        if n == 0:
            return 1
        else:
            return 0
    else:
        digit = n % 10
        if digit == 0:
            return 1 + count(n // 10)
        else:
            return count(n // 10)

num = 10203004
count(num)
```

```python
'''4. Write a python program to find Sum of Alternate Digits (Even Positions) using recursion'''
def sum(n, pos=1):
    if n == 0:
        return 0
    else:
        digit = n % 10
        if pos % 2 == 0:
            return digit + sum(n // 10, pos + 1)
        else:
            return sum(n // 10, pos + 1)

num = 123456
sum(num)
```

```python
'''5.Write a python program to find Sum of Square of Alternate Digits using recursion'''
def sum(n, pos=1):
    if n == 0:
        return 0
    else:
        digit = n % 10
        if pos % 2 == 1:
            return (digit * digit) + sum(n // 10, pos + 1)
        else:
            return sum(n // 10, pos + 1)

num = 1234
sum(num)
```

```python
'''6.Write a python program to find Count of Digits Greater than 5 using recursion'''
def count(n):
    if n == 0:
        return 0
    else:
        digit = n % 10
        if digit > 5:
            return 1 + count(n // 10)
        else:
            return count(n // 10)

num = 768495
count(num)
```
