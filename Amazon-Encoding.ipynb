```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df=pd.read_csv("Downloads/amazon_prime_video_data.csv")
```

```python
df.head()
```

## Label encoding

```python
gen=df["genre"]
encoder=LabelEncoder()
gen_encoded=encoder.fit_transform(gen)
print(gen_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
lang=df["language"]
encoder=LabelEncoder()
lang_encoded=encoder.fit_transform(lang)
print(lang_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

```python
coun=df["country"]
encoder=LabelEncoder()
coun_encoded=encoder.fit_transform(coun)
print(coun_encoded)
print("Mapping",dict(zip(encoder.classes_,range(len(encoder.classes_)))))
```

## One hot encoding

```python
df=pd.DataFrame({"gen":df["genre"]})
encoded_df=pd.get_dummies(df,columns=["gen"],dtype=int)
print(encoded_df)
```
