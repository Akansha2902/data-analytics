```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Downloads/sales_customer_analytics.csv")
```

```python
df.head()
```

```python
df.dtypes
```

```python
plt.hist(df["Quantity"])
plt.title("Quantity Distribution using Histogram")
plt.xlabel("Quantity")
plt.ylabel("Frequency")
plt.show()
```

the distribution shows many products have order quantity around 1 to 2 and there are few products with order quantity 4 followed by 3.so we can say customers usually order 1 or 2 quantities.

```python
plt.hist(df["UnitPrice"])
plt.title("Distribution of Unit Price")
plt.xlabel("Unit-Price")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows that most of the products unit prices are higher in range of 30000-40000 and 50000-60000 and users order it most of the time and lower oreder unit price range is 20000-30000.so, mostly user buy products in range of 30000-40000 and 50000-60000 in Unit price.

```python
plt.hist(df["DiscountPct"])
plt.title("Distribution of discount percentage")
plt.xlabel("Discount Percentage")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows that the most of the time discount on the products is 0.100 to 0.125 percentage and lower time 0.150 discount percentage.So, most frequently discount percentage while use.

```python
plt.hist(df["Rating"])
plt.title("Distribution of Rating")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

The distribution shows that the higher rating is 4 and lower rating is 1 as follow 2 rating of products .So, the product is good user satisfied from the product.

```python
sns.countplot(x=df["City"],data=df)
plt.title("Count Plot for City")
plt.xticks(rotation=90)
plt.show()
```

The distribution shows that the higher sale done in the chennai city follwed by pune and bangalore , lower sale in Hyderabad city with followed Mumbai city . So, The Saler should store more stocks of products in Chennai and pune city.

```python
sns.countplot(x=df["Product"],data=df)
plt.title("Count Plot for Products")
plt.xticks(rotation=90)
plt.show()
```

the distribution shows that the higher saling products is Chair and then Tablet and lowest saling product is keyboard and Headphones. So, the user can buy Chair or Tablet.

```python
sns.countplot(y=df["Category"],data=df)
plt.title("Count plot for Category")
plt.show()
```

the count plot shows that the higher selling category product is Electronics and lower selling is Furniture.So, the users usually buy Electronic products.

```python
sns.countplot(y=df["PaymentMode"],data=df)
plt.title("count plot for payment mode")
plt.show()
```

the count plot shows that the higher payment mode used by user is UPI mode as follow Cash and lower mode use is Debit Card. So, users pay usually in UPI mode.

```python
sns.countplot(x=df["OrderStatus"],data=df)
plt.title("count plot for Order status")
plt.show()
```
