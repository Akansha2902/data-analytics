```python
'''1. Write a Python program that takes a password as input and checks if it meets the following requirements:

- The password should be at least 8 characters long.
- The password should contain at least one uppercase letter.
- The password should contain at least one lowercase letter.
- The password should contain at least one digit.

If the password meets all the requirements, the program should print "Password is valid." Otherwise, it should print "Password is invalid."'''
n=input("enter a password:-")
upper=False
Lower=False
Digit=False
for i in n:
    if i.isupper():
        upper=True
    if i.islower:
        Lower=True
    if i.isdigit:
        Digit=True
if len(n)>=8 and upper and Lower and Digit:
    print("Password is valid")
else:
    print("Password is invalid")
```

```python
'''2. 
Write a Python program that takes two strings as input and checks if they're anagrams (i.e., the strings contain the same characters, but 
in a different order). If the strings are anagrams, the program should print "The strings are anagrams." Otherwise, it should print
"The strings are not anagrams."'''
n=input("enter a string:-")
m=input("enter a string:-")
if sorted(n)==sorted(m):
    print("strings are anagrams")
else:
    print("strings are not anagrams")
```
