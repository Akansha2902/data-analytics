```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Downloads/sales_customer_analytics.csv")
```

```python
df.dtypes
```

## Bivariat

```python
sns.countplot(x=df["City"],hue="Product",data=df)
plt.title("City wise Product count")
plt.xlabel("City")
plt.show()
```

The count plot shows that the highest product keyboard selling in pune city as follow desk in pune,mouse in chennai and monitor in hyderabad and lowest selling product is headphones in pune as follow desk in hyderabad. So, seller should store more stocks of these high selling products as per there city and for low selling products seller can give some discounts on it.

```python
sns.countplot(x=df["City"],hue="Category", data=df)
plt.title("count plot city wise category")
plt.xlabel("city")
plt.show()
```

The count plot shows that the highest selling category product is Electronic in Chennai city as follow in hyderabad and lowest category product selling is Furniture in mumbai.So, seller should increase stock of these electronic products in Chennai and Hyderabad and give some discount on furniture products for increasing the sells.

```python
sns.countplot(x=df["City"],hue="PaymentMode",data=df)
plt.title("Count plot for City wise Payment mode")
plt.xlabel("City")
plt.show()
```

The count plot shows that the highest payment mode UPI used in pune as followed in Bangalore,cash in chennai and lowest Debit card used in Mumbai and Hyderabad.So, seller can give more discounts as per using of cards during buying products so the user make payments through there's cards. 

```python
sns.countplot(x=df["Category"],hue="OrderStatus",data=df)
plt.title("Count plot for Category wise order status")
plt.xlabel("category")
plt.show()
```

the count plot shows that the highest delivered order is Electronics then accessories , less returned order is Furniture the Accessories and lowest cancelled order is furniture.so, seller should increasing their quality of products and solve customer problems regarding products for increasing furniture sells.
