```python
'''1.Write a python program to Convert each number into the reverse of its digits. '''
num = [123, 450, 789, 5600, 91]
reverse = lambda n, rev=0: rev if n == 0 else reverse(n // 10, rev * 10 + n % 10)
rev = list(map(reverse, num))
print(rev)
```

```python
'''Write a python program to Find the sum of the minimum and maximum in a list using reduce. '''
from functools import reduce
n=[23,4,2,56,7,90]
max=reduce(lambda x,y:x if x>y else y,n)
min=reduce(lambda x,y:x if x<y else y,n)
print(max+min)
```

```python
'''6.Write a python program to Count how many numbers in a list are greater than the average (use reduce to get average). '''
from functools import reduce
n = [23, 4, 2, 56, 7, 90]
total = reduce(lambda a, b: a + b, n)
count_elements = reduce(lambda count, element: count + 1, n, 0)
avg = total / count_elements
count= reduce(lambda x, num: x + 1 if num > avg else x, n, 0)
print(f"count = {count}, average = {avg:.2f}")
```

```python
'''Write a python program to Filter numbers whose digit sum is even, then return their reversed forms. '''
from functools import reduce

n = [12, 34, 45, 23, 56, 7, 89, 101]
add= lambda x: 0 if x == 0 else x % 10 + add(x // 10)
reverse = lambda n, rev=0: rev if n == 0 else reverse(n // 10, rev * 10 + n % 10)
filtered = filter(lambda x: add(x) % 2 == 0, n)
reversed_numbers = list(map(reverse, filtered))
print(reversed_numbers)
```
