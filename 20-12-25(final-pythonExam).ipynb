```python
#15
a=[2,6,8,9,10,100,99]
b=[10,2,8,5,6,11,34]
c=[]
a1=[]
b1=[]
for i in a:
    if i in b:
        c+=[i]
    else:
        a1+=[i]
for j in b:
    if j not in c:
        b1+=[j]
print(f"common element in both list ={c}")
print(f"non-common element in a={a1} and in b={b1}")
```
