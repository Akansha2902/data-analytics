```python
'''1.Write a function create_student_profile(name, *scores, **weights) that:
Accepts name as the first argument.
Accepts any number of scores in *scores (assume test scores).
Accepts keyword arguments as subject weights (e.g., math=2, english=1).
Returns a dictionary'''
def create_student_profile(name, *scores, **weights):
    print(name)
    print("scores:-")
    for i in scores:
      print(i)
    print("weights:-")
    for k,v in weights.items():
      print(k,v)
create_student_profile("akansha",23,34,21,45,w1=34,w2=43)
```

```python
'''5. Write a function check_primes(*args) that takes multiple integers and returns a dictionary where each number is a key, and the value is True if it's
a prime number, otherwise False.'''
def check_primes(*args):
    def prime(n):
        if n < 2:
            return False
        for i in range(2, int(n**0.5) + 1):
            if n % i == 0:
                return False
        return True

    return {n: prime(n) for n in args}

check_primes(23,56,78,12,13,17)
```

```python
'''6. Write a function check_perfect_numbers(*args) that:
Accepts multiple integers via *args
Returns a list of only those numbers that are perfect numbers'''
def check_perfect_numbers(*args):
    perfects = []
    for n in args:
        if n > 1:
            d = [i for i in range(1, n) if n % i == 0]
            if sum(d) == n:
                perfects.append(n)
    return perfects

check_perfect_numbers(34,567,879,28,120,234,564,6)
```

```python
'''7. Write a function filter_divisible(divisor, *args) that returns a list of numbers from args divisible by divisor.
'''
def filter_divisible(divisor, *args):
    if divisor == 0:
        return("Divisor cannot be zero.")

    result = []
    for n in args:
        if n % divisor == 0:
            result.append(n)
    return result
filter_divisible(2, 1, 2, 3, 4, 5, 6)

```

```python
'''8.Write a function is_prime(num) that returns True if num is prime, else False.'''
def is_prime(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

is_prime(13)
```

```python
'''9.Write a function count_divisors(*args) that returns a dictionary mapping each number to the count of its positive divisors.'''
def count_divisors(*args):
    result = {}
    for n in args:
        count = 0
        for i in range(1, n + 1):
            if n % i == 0:
                count += 1
        result[n] = count
    return result

count_divisors(34,67,89,23,13)
```

```python
'''
10.Write a function sum_of_powers(power, *args) that computes the sum of each number raised to the given power (power is a positive integer).'''
def sum_of_powers(power, *args):
    if power <= 0:
       return ("Power must be a positive integer.")

    total = 0
    for n in args:
        total += n ** power
    return total
print(sum_of_powers(2, 1, 2, 3, 4))
```
