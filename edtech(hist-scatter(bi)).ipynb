```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Downloads/edtech_student_learning_data.csv")
```

```python
df.dtypes
```

## Histogram

```python
plt.hist(df["total_sessions"])
plt.title("Distribution of total sessions")
plt.xlabel("total sessions")
plt.ylabel("Frequency")
plt.show()
```

The histogram shows that the there is total 500 sessions.

```python
plt.hist(df["sessions_attended"])
plt.title("Distribution of sessions_attendede")
plt.xlabel("sessions_attended")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest session attended in range 60-70 and lowest in range 110-120.

```python
plt.hist(df["avg_quiz_score"])
plt.title("Distribution of average quiz score")
plt.xlabel("average quiz score")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest average quiz score is in range 55-65 and lowest in range 80-90.

```python
plt.hist(df["assignment_score"])
plt.title("Distribution of assignment score")
plt.xlabel("assignment score")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest assignment score in range 40-50 as follow 60-70 and lowest in range 50-60.

```python
plt.hist(df["project_score"])
plt.title("Distribution of project score")
plt.xlabel("project score")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest appearing project score in range 80-100 and lowest in range 55-60.

```python
plt.hist(df["learning_hours"])
plt.title("Distribution of learning hours")
plt.xlabel("learning hours")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest appearing learning hours in range 10-15 and lowest in range 5-10.

```python
plt.hist(df["salary_offered"])
plt.title("Distribution of salary_offered")
plt.xlabel("salary offered")
plt.ylabel("Frequency")
plt.show()
```

the histogram shows that the highest appearing salary offered in range 1-1.2 and lowest in range 0.6.
